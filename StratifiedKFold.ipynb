```python
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, LeaveOneOut, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
x= df.drop(["species"], axis=1)
y= df[["species"]]
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x,y, random_state=42, test_size=.2)
```

```python
model=dt.fit(xtrain, ytrain)
model.score(xtrain, ytrain)
```

```python
pred=model.predict(xtest)
cv=accuracy_score(ytest, pred)
cv
```

```python
strd=StratifiedKFold()
```

```python
scor1=cross_val_score(dt, x,y, cv=strd)
scor1
```
